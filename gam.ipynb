```python
#GAM
import pandas as pd
data_gejala = pd.read_excel (r'nilai_gam.xlsx')
data_gejala.tail(38)
```

```python
#mengisi nilai inferensi

inferensi = {
    "tinggi":100,
    "rendah":10
}
```

```python
data_pasien = pd.read_excel (r'data_pasien.xlsx')
data_pasien.head(10)
```

```python
# import fungsi
def turun(b,a,x):
    if(x<=a):
        nilai = 1
    elif(x>a and x<b):
        nilai = (b-x)/(b-a)
    elif(x>=b):
        nilai = 0
    return nilai

def naik(b,a,x):
    if(x<=a):
        nilai = 0
    elif(x>a and x<b):
        nilai = (x-a)/(b-a)
    elif(x>=b):
        nilai = 1
    return nilai

def agregasi_turun(b,a,alfa):
    nilai = b - (alfa*(b-a))
    return nilai

def agregasi_naik(b,a,alfa):
    nilai = alfa*(b-a) + a
    return nilai
```

```python
#merge kolom gam dan data pasien

gabung_data = pd.merge(data_gejala,data_pasien[['no','nilai']],how='inner', on='no')
gabung_data.columns = ['no','kode','gejala','tinggi','rendah','nilai']
gabung_data.tail(5)
```

```python
#memasukan nilai naik dan nilai turun

gabung_data['nilai_naik'] = gabung_data.apply(lambda x : naik(x["tinggi"],x["rendah"],x["nilai"]),axis=1)
gabung_data['nilai_turun'] = gabung_data.apply(lambda x : turun(x["tinggi"],x["rendah"],x["nilai"]),axis=1)

gabung_data.head(10)
```

```python
# melakukan input rules

data_rule = pd.read_excel (r'rule_gam.xlsx')
data_rule.head(10)
```

```python
# melihat jumlah rule yang aktif
data_columns_rule = list(data_rule.columns)

gangguan = data_columns_rule.pop()
data_columns_rule

gangguan
```

```python
z = []
alfa = []
for index, row in data_rule.iterrows():
    hasil_per_row = []
    for per_g in data_columns_rule:
        fetching_data_gabung = gabung_data.loc[gabung_data["kode"]== per_g]
        if(row[per_g] == "R"):
            data = fetching_data_gabung['nilai_turun'].to_string(header=False, index=False).replace(' ','')
        else:
            data = fetching_data_gabung['nilai_naik'].to_string(header=False, index=False).replace(' ','')
        hasil_per_row.append(data)
        
#     print(hasil_per_row)
    a = min(hasil_per_row)
    alfa.append(float(a))
    if(data_rule[gangguan][index] == "R"): #jika kesimpulan turun atau rendah
        zz = agregasi_turun(float(inferensi['tinggi']),float(inferensi['rendah']),float(a))
    elif(data_rule[gangguan][index] == "T"): #jika kesimpulan tinggi atau naik
        zz = agregasi_naik(float(inferensi['tinggi']),float(inferensi['rendah']),float(a))        
    z.append(zz)
#     break
```

```python
print(alfa)
print(z)
```

```python
#DEFUZIFIKASI
df = 0
for i in range(len(alfa)):
    df += alfa[i]*z[i]
defuz = int(df/sum(alfa))
print("Jadi, nilai",gangguan,"adalah ",defuz)
```
